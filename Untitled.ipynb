```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit

# Load the Excel file
df = pd.read_excel(r"C:\Users\Lenovo\Desktop\az.xlsx")

# Calculate L2-R and L2-N
df['L2-R'] = 100 - df['L1-R']
df['L2-N'] = 100 - df['L1-N']

# Calculate Rx for both sets using given R values
df['Rx-R'] = 100 * df['L1-R'] / df['L2-R']
df['Rx-N'] = 1000 * df['L1-N'] / df['L2-N']

# Print the calculated values with one decimal place
print("L2-R values:", [f"{x:.1f}" for x in df['L2-R']])
print("L2-N values:", [f"{x:.1f}" for x in df['L2-N']])
print("Rx-R values:", [f"{x:.1f}" for x in df['Rx-R']])
print("Rx-N values:", [f"{x:.1f}" for x in df['Rx-N']])

# Plot the results
plt.figure(figsize=(10, 6))

# Linear regression for set 1 (L1-R, L2-R)
X1 = df[['temp']]
y1 = df['Rx-R']
linear_model = LinearRegression()
linear_model.fit(X1, y1)
y1_pred = linear_model.predict(X1)
plt.plot(df['temp'], y1, 'bo')
plt.plot(df['temp'], y1_pred, 'b--', label='Conductor')

# Calculate error for set 1
error1 = np.mean((y1 - y1_pred)**2)
plt.text(0.5, 0.27, f"Conductor: y = {linear_model.coef_[0]:.2f}x + {linear_model.intercept_:.2f}\nError: {error1:.2f}", 
         transform=plt.gca().transAxes, fontsize=10, color='blue')

# Exponential regression for set 2 (L1-N, L2-N)
def exp_func(x, a, b):
    return a * np.exp(b * x)

X2 = df['temp']
y2 = df['Rx-N']
popt, _ = curve_fit(exp_func, X2, y2, p0=(1, 0.01))
y2_pred = exp_func(X2, *popt)
plt.plot(df['temp'], y2, 'ro')
plt.plot(df['temp'], y2_pred, 'r:', label='SemiConductor')

# Calculate error for set 2
error2 = np.mean((y2 - y2_pred)**2)
plt.text(0.15, 0.8, f"SemiConductor: y = {popt[0]:.2f} * exp({popt[1]:.2f}x)\nError: {error2:.2f}",
         transform=plt.gca().transAxes, fontsize=10, color='red')

# Label the axes and the plot
plt.xlabel('Temperature(C)')
plt.ylabel('Resistance ()')
plt.legend()
plt.title('Resistance vs Temperature for Conductor and SemiConductor')

# Save the plot as a file
plt.savefig(r"C:\Users\Lenovo\Desktop\Resistance_vs_Temperature.png")

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit

# Load the Excel file
df = pd.read_excel(r"C:\Users\Lenovo\Desktop\az.xlsx")

# Calculate L2-R and L2-N
df['L2-R'] = 100 - df['L1-R']
df['L2-N'] = 100 - df['L1-N']

# Calculate Rx for both sets using given R values
df['Rx-R'] = 100 * df['L1-R'] / df['L2-R']
df['Rx-N'] = 1000 * df['L1-N'] / df['L2-N']

# Print the calculated values with one decimal place
print("L2-R values:", [f"{x:.1f}" for x in df['L2-R']])
print("L2-N values:", [f"{x:.1f}" for x in df['L2-N']])
print("Rx-R values:", [f"{x:.1f}" for x in df['Rx-R']])
print("Rx-N values:", [f"{x:.1f}" for x in df['Rx-N']])

# Save the calculated data to a new Excel file
df.to_excel(r"C:\Users\Lenovo\Desktop\calculated_data.xlsx", index=False)

# Plot the results
plt.figure(figsize=(10, 6))

# Linear regression for set 1 (L1-R, L2-R)
X1 = df[['temp']]
y1 = df['Rx-R']
linear_model = LinearRegression()
linear_model.fit(X1, y1)
y1_pred = linear_model.predict(X1)
plt.plot(df['temp'], y1, 'bo', label='Actual Rx-R')
plt.plot(df['temp'], y1_pred, 'b-', label='Fitted Rx-R (Linear)')

# Calculate error for set 1
error1 = np.mean((y1 - y1_pred)**2)
plt.text(0.9, 0.9, f"Conductor: y = {linear_model.coef_[0]:.2f}x + {linear_model.intercept_:.2f}\nError: {error1:.2f}", 
         transform=plt.gca().transAxes, fontsize=10, color='blue')

# Exponential regression for set 2 (L1-N, L2-N)
def exp_func(x, a, b):
    return a * np.exp(b * x)

X2 = df['temp']
y2 = df['Rx-N']
popt, _ = curve_fit(exp_func, X2, y2, p0=(1, 0.01))
y2_pred = exp_func(X2, *popt)
plt.plot(df['temp'], y2, 'ro', label='Actual Rx-N')
plt.plot(df['temp'], y2_pred, 'r:', label='Fitted Rx-N (Exponential)')

# Calculate error for set 2
error2 = np.mean((y2 - y2_pred)**2)
plt.text(0.9, 0.8, f"SemiConductor: y = {popt[0]:.2f} * exp({popt[1]:.2f}x)\nError: {error2:.2f}",
         transform=plt.gca().transAxes, fontsize=10, color='red')

# Label the axes and the plot
plt.xlabel('Temperature(C)')
plt.ylabel('Resistance ()')
plt.legend()
plt.title('Resistance vs Temperature for Conductor and SemiConductor')

# Save the plot as a file
plt.savefig(r"C:\Users\Lenovo\Desktop\Resistance_vs_Temperature.png")

# Show the plot
plt.show()
```
